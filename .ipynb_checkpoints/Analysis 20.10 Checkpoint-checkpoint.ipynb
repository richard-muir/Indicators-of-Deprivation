```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
%matplotlib inline
```

```python
deprivation_xl = pd.ExcelFile('Deprivation.xls')
dep_df = deprivation_xl.parse('IMD 2010')
dep_df.rename(columns={'IMD SCORE':'DEPRIVATION SCORE'},inplace=True)
dep_df=DataFrame(dep_df)
dep_df.drop('RANK OF IMD SCORE (where 1 is most deprived)', 
            axis=1, inplace=True)
dep_df
```

```python
age_xl = pd.ExcelFile('Age.xls')
age_df = age_xl.parse('QS103EW_Numbers',header=10)
age_df.rename(columns={'Area code':'GOR'},inplace=True)
age_df = DataFrame(age_df)
age_df.head()
```

```python
xage_df = age_df.query("""GOR in ['E12000001','E12000002','E12000003','E12000004','E12000005','E12000006','E12000007','E12000008','E12000009']""")
```

```python
yage_df = xage_df.drop(['Unnamed: 2','Unnamed: 3'],axis=1)
```

```python
ages = list(yage_df.loc[:])
```

```python
adult = ages[21:69]
```

```python
child = ages[3:21]
```

```python
pensioner=ages[69:104]
```

```python
yage_df['Children']=yage_df[child].sum(axis=1)
```

```python
yage_df['Adults']=yage_df[adult].sum(axis=1)
```

```python
yage_df['Pensioners']=yage_df[pensioner].sum(axis=1)
```

```python
yage_df.head()
```

```python
for i in range(len(child)):
    yage_df.drop([child[i]], axis=1, inplace=True)
       
```

```python
for j in range(len(adult)):
    yage_df.drop([adult[j]], inplace=True, axis=1)
    
            
```

```python
for k in range(len(pensioner)):
    yage_df.drop([pensioner[k]], inplace=True, axis=1)
```

```python
# checking that everything was summed correctly
col_list = list(yage_df)
col_list = col_list[3:]
col_list
zage_df = yage_df[col_list].sum(axis=1)
```

```python
communal_xl = pd.ExcelFile('Communal Living.xls')
communal_df = communal_xl.parse('QS101EW_Numbers', header=10)
communal_df.rename(columns={'Area code':'GOR'},inplace=True)
communal_df = DataFrame(communal_df)
communal_df.head()
```

```python
eth_xl = pd.read_csv('EthnicityGOR.csv', header=10)
eth_df = DataFrame(eth_xl)
eth_df.rename(columns={'Unnamed: 0':'GOR', 'Unnamed: 1':'Area name'}, inplace=True)
eth_df = eth_df.query("""GOR in ['E12000001','E12000002','E12000003','E12000004','E12000005','E12000006','E12000007','E12000008','E12000009']""")
eth_df = eth_df.drop(['Unnamed: 2'],axis=1)
eth_df.head()
```

```python
# combine the columns with less than 1000 obs into other
```

```python
orig_col_list = list(eth_df)
num_col_list = orig_col_list[2:]
sum_col_list = orig_col_list[3:-1]
num_col_list
```

```python
for col in num_col_list:
    if eth_df[col].dtype == object:
        eth_df[col] = eth_df[col].str.replace(',','')
```

```python
sum_col_list
```

```python
eth_df[num_col_list] = eth_df[num_col_list].convert_objects(convert_numeric=True).astype('float32')
eth_df
```

```python
for col in sum_col_list:
    if eth_df[col].sum() < 250000:
        print(str(eth_df[col]), eth_df[col].sum())
        print("other before sum: " + str(eth_df[num_col_list[-1]]))
        eth_df[num_col_list[-1]] += eth_df[col]
        print("other after sum: " + str(eth_df[num_col_list[-1]]))
        eth_df.drop([col], axis=1, inplace=True)
        
```

```python
eth_df.head()
```

```python
new_col_list = eth_df.columns.values.tolist()
new_col_list = new_col_list[3:]
len(new_col_list)
```

```python
sum_test = eth_df[new_col_list].sum(axis=1)
sum_test
```

```python
rel_xl = pd.read_csv('ReligionGOR.csv', header=10)
rel_df = DataFrame(rel_xl)
rel_df.rename(columns={'Unnamed: 0':'GOR', 'Unnamed: 1':'Area name'}, inplace=True)
rel_df = rel_df.query("""GOR in ['E12000001','E12000002','E12000003','E12000004','E12000005','E12000006','E12000007','E12000008','E12000009']""")
rel_df.drop(['Unnamed: 2'], axis=1, inplace=True)
rel_df
```

```python
rel_orig_col_list = list(rel_df)
rel_sum_col_list = rel_orig_col_list[9:]
rel_sum_col_list
```

```python
for col in rel_sum_col_list:
    if col not in ['Other religion: Total','No religion: Total','Religion not stated']:
        rel_df.drop([col], axis=1, inplace=True)
```

```python
dens_xl = pd.ExcelFile('Population Density.xls')
dens_df = dens_xl.parse('QS102EW_Numbers', header=10)
dens_df = DataFrame(dens_df)
dens_df.rename(columns={'Area code':'GOR','Area':'Area in Hectares','Density':'Density (Persons per hectare)'}, inplace=True)
dens_df = dens_df.query("""GOR in ['E12000001','E12000002','E12000003','E12000004','E12000005','E12000006','E12000007','E12000008','E12000009']""")
dens_df = dens_df.drop(['Unnamed: 2','Unnamed: 3'], axis=1)
dens_df
```

```python
#get average or total deprivation score for GORs
```

```python
mean_dep_df = dens_df.loc[:,['GOR','Area name']]
```

```python
mean_dep_df['Deprivation'] = 1
```

```python
mean_dep_df.index = range(9)
mean_dep_df
```

```python
areas = list(mean_dep_df['Area name'])
areas[5] = 'EAST OF ENGLAND'
areas[1] = areas[1].strip()
areas
```

```python
LSOA = list(dep_df['LSOA CODE'])
```

```python
means=[]
for i in range(len(areas)):
    for j in range(len(LSOA)):
        if dep_df.loc[j,'GOR NAME'].upper() == areas[i]:
            means.append(dep_df.loc[j,'DEPRIVATION SCORE'])
    mean_dep_df.loc[i,'Deprivation'] = np.mean(means)
    means=[]
```
